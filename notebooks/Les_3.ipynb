N-gram tekst generatie:

```python
import sys
sys.path.append('..')

from ngram import *
from nlp import save_tokens
```

Het ngram model zal hier een context grootte van 3 tokens gebruiken en 100 tokens te genereren.
Met de tokenizer moet eerst een *.enc bestand met de tokeniser gemaakt worden. 

```python
vocabulary_list, story_token_list = file_reader("../ongebruikte_scripts/encoding.enc")

model = train_ngram_model(story_token_list, n=3)

# Genereer tekst:
generated_text = generate_text(model, n=3, length=100)

# Sla de gegenereerde tekst op in een .enc bestand:
save_tokens("generated_tokens_ngram.enc", generated_text)
```

```python
print(generated_text)
```

Met ander N groottes:

```python
model = train_ngram_model(story_token_list, n=2)
generated_text = generate_text(model, n=2, length=100)
print(generated_text)
```

```python
model = train_ngram_model(story_token_list, n=50)
generated_text = generate_text(model, n=50, length=100)
print(generated_text)
```

**Interpretatie:**

N=1: De tekst is volledig willekeurig. Woorden hebben geen relatie met hun buren.

N=2: We zien lokale samenhang (bijv. "of the", "in a"). De zinnen lopen echter niet logisch door.

N=3: De tekst begint op echt Engels te lijken. Zinsdelen zijn vaak grammaticaal correct, maar over langere afstanden slaat de inhoud nergens op. 

Een hogere N levert geloofwaardiger Engels op qua structuur, maar het model kopieert ook grotere letterlijke stukken uit de brontekst (overfitting op data). Met een N die hoog genoeg is zou het model dezelfde tekst produceren die gebruikt was voor het trainen.
