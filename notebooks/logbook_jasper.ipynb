# Data inladen

```python
with open("../kanker_nl.txt", "r", encoding="utf-8") as f:
    nl_text = f.read()

with open("../cancer_en.txt", "r", encoding="utf-8") as f:
    en_text = f.read()

print(nl_text[:500])
print("\nEngelse versie\n")
print(en_text[:500])
```

Hier laat ik tokenize trainen op de Nederlands data

```python
%%bash
python tokenize_jasper.py learn \
    --text ../kanker_nl.txt \
    --enc kanker_nl.enc \
    --tokens kanker_nl.tok \
    --max-tokens 5000 \
    --min-freq 3 \
    --lowercase
```

Hier laar ik het trainen op de engelse data

```python
%%bash
python tokenize_jasper.py learn \
    --text ../cancer_en.txt \
    --enc cancer_en.enc \
    --tokens cancer_en.tok \
    --max-tokens 5000 \
    --min-freq 3 \
    --lowercase
```

gebruik de nederlandse training en de engelse token data set op de nedeerlandse tekst

encode breekt een tekst op in stukjes die de tokenizer kent,
zet die stukjes om in nummers,

en slaat ze op.

```python
%%bash
python tokenize_jasper.py encode \
    --text ../kanker_nl.txt \
    --enc kanker_nl.enc \
    --tokens kanker_nl_on_nl.tok

python tokenize_jasper.py encode \
    --text ../kanker_nl.txt \
    --enc cancer_en.enc \
    --tokens kanker_nl_on_en.tok
```

gebruik de nederlandse training en de engelse token data set op de 
engelse  tekst

encode breekt een tekst op in stukjes die de tokenizer kent,
zet die stukjes om in nummers,

en slaat ze op.

```python
%%bash
python tokenize_jasper.py encode \
    --text ../cancer_en.txt \
    --enc kanker_nl.enc \
    --tokens cancer_en_on_nl.tok

python tokenize_jasper.py encode \
    --text ../cancer_en.txt \
    --enc cancer_en.enc \
    --tokens cancer_en_with_nl_enc.tok
```

Nu van de tokens weer een tekst maken 

```python
%%bash
python tokenize_jasper.py decode \
    --tokens kanker_nl_on_nl.tok \
    --enc kanker_nl.enc \
    --text ../reconstructed_nl.txt
```

```python
with open("../kanker_nl.txt", "r", encoding="utf-8") as f:
    nl_text = f.read()

with open("../reconstructed_nl.txt", "r", encoding="utf-8") as f:
    decode_text = f.read()

print(nl_text[:500])
print("\ndecode\n")
print(decode_text[:500])
```
